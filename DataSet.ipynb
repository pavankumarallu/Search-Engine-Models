```python
import pandas as pd
```

```python
CeraFloor = pd.read_csv("CeraFloorTiles.csv").dropna()
CeraWall = pd.read_csv("CeraWallTiles.csv",usecols=range(4)).dropna()
OrientTiles = pd.read_csv("OrientTiles.csv",usecols=range(4)).dropna()
RAK = pd.read_csv("RAK.csv").dropna()
Johnson = pd.read_csv("johnson.csv").dropna()
mytiles = pd.read_csv("mytyles.csv",usecols=range(4)).dropna()
```

```python
NitcoFloor = pd.read_csv('NitcoFloorTiles.csv',usecols=range(4),encoding='latin-1').dropna()
```

```python
Companies = [CeraFloor,CeraWall,OrientTiles,RAK,Johnson,mytiles,NitcoFloor]
```

```python
CeraFloor['Training Images'] = CeraFloor['Image']
CeraFloor['Company'] = 'CeraFLoors'
CeraWall['Company'] = 'CeraWall'
OrientTiles.rename( columns={'Unnamed: 3':'Training Images','Images':'Image'}, inplace=True )
OrientTiles['Company'] = 'Orient Tiles'
RAK['Training Images'] = RAK['Image']
RAK['Company'] = 'RAK'
RAK['Link'] = RAK['Image']
Johnson['Training Images'] = Johnson['Image']
Johnson['Company'] = 'Johnson'
mytiles.rename( columns={'image-0':'Training Images','Main-Image':'Image'}, inplace=True )
mytiles['Company'] = 'My tiles'
NitcoFloor['Company'] = 'Nitco FLoor'
```

```python
Final = pd.concat(Companies,ignore_index=True)
Final
```

```python
Final.to_csv('final.csv')
```

```python
import urllib.request
```

```python
image_data = {}
count = 1
for i in range(len(Final)):
  try:
    image_url = Final['Training Images'][i]
    filename = '/content/drive/MyDrive/ImageDataSetTiles/'+Final['Company'][i]+'_'+str(count)+".jpg"
    img_dict = {
        'Company' : Final['Company'][i] , 
        'Name' : Final['Name'][i],
        'Link' : Final['Link'][i]
    }
    image_data[Final['Company'][i]+'_'+str(count)+".jpg"] = img_dict
    urllib.request.urlretrieve(image_url, filename)
    print(filename)
    count = count+1
  except:
    pass
  
```

```python
import json
with open("ImageData.json", "w") as outfile:
    json.dump(image_data, outfile)
```
